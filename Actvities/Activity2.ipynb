```python
# Contour-Based Edge Detection in Google Colab

import cv2
import numpy as np
from google.colab.patches import cv2_imshow

# -------------------------------
# 1. Load the image
# -------------------------------

image_path = "/content/tiger.jpeg"
image = cv2.imread(image_path)

if image is None:
    print("Error: Image not found! Check the file path.")
    exit()

print("Original image shape:", image.shape)

# -------------------------------
# 2. Convert to grayscale
# -------------------------------

gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# -------------------------------
# 3. Apply Gaussian blur (optional, reduces noise)
# -------------------------------

blur = cv2.GaussianBlur(gray, (5, 5), 0)

# -------------------------------
# 4. Canny edge detection
# -------------------------------
edges = cv2.Canny(blur, threshold1=50, threshold2=150)

# -------------------------------
# 5. Find contours
# -------------------------------
contours, hierarchy = cv2.findContours(edges, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

# -------------------------------
# 6. Draw contours on a copy of the original image
# -------------------------------
contour_image = image.copy()
cv2.drawContours(contour_image, contours, -1, (0, 255, 0), 2)  # green contours

# -------------------------------
# 7. Display results in Colab
# -------------------------------

print("Original Image")
cv2_imshow(image)

print("Canny Edges")
cv2_imshow(edges)

print("Contours on Image")
cv2_imshow(contour_image)

# -------------------------------
# 8. Save the results
# -------------------------------

cv2.imwrite("/content/edges_canny.png", edges)
cv2.imwrite("/content/contours_image.png", contour_image)

print("Saved edges_canny.png and contours_image.png in Colab")
```
